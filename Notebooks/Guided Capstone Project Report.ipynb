## Introduction<a id='Introduction'></a>

The purpose of this report is to highlight the results of data analysis and modelling for estimating a pricing model for ski resort tickets in Big Mountain's market segment. Big Mountain suspected that it may not be maximizing its returns, relative to its position in the market. There was also not a strong sense of what facilities matter most to visitors, particularly which ones they're most likely to pay more for. This results below demonstrate a predictive model to provide guidance for Big Mountain's pricing and future facility investment plans.

## Data<a id='Introduction'></a>

The original data supplied had 330 rows and 27 columns of data, with all data available for the Big Mountain resort. After some data wrangling, a few rows and columns were removed beacuse of missing price data or inaccurate values. The adult weekend ticket price was chosen as our main pricing model as it had the least missing data.

Using the scatterplots above, we were able to determine positive correlations for fastQuads, runs, snow making area in acres, and vertical drops. We used these preliminary correlations as points of interest when determining our machine learning model.

I first split up our data to train the models. The mean did slightly worse on the test set than on the training set. Using the mean absolute error, I found that ticket price would be around \\$19 off if one were to guess ticket price based on an average of known values.

A linear model explained over 80\% of the variance on the train set and over 70\% on the test set. The lower value from the test set suggests that I overfit the model. Using just the linear model, I would be able to estimate the ticket price within \\$9 of the real price. I then refined the linear model by fitting it to a new pipline. I found that for this pipline, a k value of 8 worked the best. This model showed taht vertical drop is the park's largest positive feature. I then used a random forest model, shown above, that predicted that vertical drop would be in the top 4 features, with fastQuads being the highest feature for this model. The random forest model had a lower crosss-validation mean absolute error than the linear regression model by around \\$1, and had less variabiltiy. I decided to use the random forest model moving forward with the machine learning algorithm.
